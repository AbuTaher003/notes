```python
# Question:
# Simulate a Manufacturing System.

import random
import matplotlib.pyplot as plt

# Simulation parameters
simulation_time = 50

# Processing time distribution
processing_time = [2, 3, 4, 5]
probability = [0.2, 0.4, 0.3, 0.1]

time = 0
jobs_completed = 0

completion_times = []

# Simulation
while time < simulation_time:

    process = random.choices(
        processing_time,
        weights=probability
    )[0]

    time += process

    if time <= simulation_time:
        jobs_completed += 1
        completion_times.append(jobs_completed)

print("Simulation Time:", simulation_time)
print("Jobs Completed:", jobs_completed)

# Visualization
plt.plot(range(1, jobs_completed + 1),
         completion_times,
         marker='o')

plt.xlabel("Job Number")
plt.ylabel("Completed Jobs")
plt.title("Manufacturing System Simulation")
plt.grid(True)

plt.show()
```
