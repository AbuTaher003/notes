```python
# Question:
# Implement the Poker Test to assess the randomness
# of numbers generated by a Multiplicative Congruential Generator (MCG).

# Multiplicative Congruential Generator (MCG)
a = 5
m = 97
seed = 7
n = 20

random_numbers = []

# Generate random numbers
x = seed
for i in range(n):
    x = (a * x) % m
    random_numbers.append(x)

print("Generated Numbers:")
print(random_numbers)

# Poker Test
all_same = 0
one_pair = 0
all_different = 0

for num in random_numbers:
    # Make every number 4 digits
    s = str(num).zfill(4)

    # Count frequency of each digit
    freq = {}
    for digit in s:
        freq[digit] = freq.get(digit, 0) + 1

    counts = list(freq.values())

    if len(freq) == 1:
        all_same += 1          # AAAA
    elif 2 in counts:
        one_pair += 1          # AABC or AABB
    elif len(freq) == 4:
        all_different += 1     # ABCD

print("\nPoker Test Result:")
print("All Same (AAAA):", all_same)
print("One Pair :", one_pair)
print("All Different :", all_different)

# Expected probabilities
p_all_same = 0.001
p_one_pair = 0.504
p_all_diff = 0.495

e1 = n * p_all_same
e2 = n * p_one_pair
e3 = n * p_all_diff

# Chi-Square Calculation
chi_square = ((all_same - e1) ** 2) / e1
chi_square += ((one_pair - e2) ** 2) / e2
chi_square += ((all_different - e3) ** 2) / e3

print("\nChi-Square Value =", round(chi_square, 4))

# Critical value (df=2, alpha=0.05)
critical = 5.991

if chi_square < critical:
    print("Accept H0 (Random Numbers)")
else:
    print("Reject H0 (Not Random)")
```
