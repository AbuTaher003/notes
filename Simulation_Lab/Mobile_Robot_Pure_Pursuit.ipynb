```python
# Question:
# You are tasked with simulating a mobile robot that follows a predefined path
# using the Pure Pursuit algorithm. The robot should start at an initial position
# and navigate towards a sequence of waypoints using the pure pursuit strategy.
# Your solution must include a real-time animation of the robot moving along the path.

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Predefined path (waypoints)
path_x = [0, 2, 4, 6, 8, 10]
path_y = [0, 1, 0, 2, 1, 0]

# Robot initial position
robot_x = path_x[0]
robot_y = path_y[0]

# Store robot trajectory
traj_x = [robot_x]
traj_y = [robot_y]

# Create figure
fig, ax = plt.subplots()
ax.plot(path_x, path_y, 'ro--', label='Path')
robot, = ax.plot([], [], 'bo', markersize=10, label='Robot')
trajectory, = ax.plot([], [], 'b-', label='Trajectory')

ax.set_xlim(-1, 11)
ax.set_ylim(-1, 3)
ax.legend()
ax.grid()

# Animation function
def update(frame):
    global robot_x, robot_y

    # Current target waypoint
    target_x = path_x[frame]
    target_y = path_y[frame]

    # Move robot toward target (Pure Pursuit concept)
    robot_x = target_x
    robot_y = target_y

    traj_x.append(robot_x)
    traj_y.append(robot_y)

    robot.set_data([robot_x], [robot_y])
    trajectory.set_data(traj_x, traj_y)

    return robot, trajectory

# Real-time animation
ani = FuncAnimation(
    fig,
    update,
    frames=len(path_x),
    interval=1000,
    repeat=False
)

plt.title("Pure Pursuit Robot Simulation")
plt.show()
```
