```python
# Question:
# You are tasked with analyzing the randomness of a sequence of numbers
# generated using the Additive Congruential Method (ACM).
# To evaluate the independence of these random numbers,
# you will implement a Chi-Square test for autocorrelation.

# Additive Congruential Method parameters
m = 100
x1 = 23
x2 = 45
n = 20

random_numbers = [x1, x2]

# Generate random numbers using ACM
for i in range(n - 2):
    x = (random_numbers[-1] + random_numbers[-2]) % m
    random_numbers.append(x)

print("Generated Numbers:")
print(random_numbers)

# Convert to Uniform(0,1)
u = []
for x in random_numbers:
    u.append(x / m)

# Autocorrelation pairs
observed = [[0, 0], [0, 0]]

for i in range(len(u) - 1):
    a = 0 if u[i] < 0.5 else 1
    b = 0 if u[i + 1] < 0.5 else 1
    observed[a][b] += 1

print("\nObserved Frequency Matrix:")
for row in observed:
    print(row)

# Expected frequency
total = len(u) - 1
expected = total / 4

chi_square = 0

print("\nCell\tObserved\tExpected\t(O-E)^2/E")

for i in range(2):
    for j in range(2):
        value = ((observed[i][j] - expected) ** 2) / expected
        chi_square += value

        print(f"{i}{j}\t",
              observed[i][j], "\t\t",
              expected, "\t\t",
              round(value, 4))

print("\nChi-Square Value =", round(chi_square, 4))

# Critical value for df = 3 and alpha = 0.05
critical = 7.815

if chi_square < critical:
    print("Accept H0 (Independent Random Numbers)")
else:
    print("Reject H0 (Dependent Random Numbers)")
    
```
