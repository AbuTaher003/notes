```python
# Question:
# You are tasked with simulating the chemical reaction between hydrogen (H2)
# and oxygen (O2) to form water (H2O).
#
# Balanced equation:
# 2H2 + O2 -> 2H2O
#
# Given an initial amount of hydrogen (h2_molecules) and oxygen
# (o2_molecules), simulate the reaction until no more water can be formed.
#
# Output the remaining H2, O2, and the amount of H2O formed.

# Initial molecules
h2_molecules = 10
o2_molecules = 5

# Water formed
h2o_molecules = 0

# Simulate reaction
while h2_molecules >= 2 and o2_molecules >= 1:
    h2_molecules -= 2
    o2_molecules -= 1
    h2o_molecules += 2

# Output
print("Equilibrium State:")
print("Remaining H2 =", h2_molecules)
print("Remaining O2 =", o2_molecules)
print("H2O Formed   =", h2o_molecules)
```
