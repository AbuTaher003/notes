```python
# Question:
# Simulate a Profit Analysis Problem.

import random
import matplotlib.pyplot as plt

# Parameters
simulation_days = 30

selling_price = 100
production_cost = 60

daily_profit = []

total_profit = 0

# Simulation
for day in range(1, simulation_days + 1):

    # Random daily sales
    units_sold = random.randint(20, 50)

    revenue = units_sold * selling_price
    cost = units_sold * production_cost

    profit = revenue - cost

    daily_profit.append(profit)
    total_profit += profit

print("Total Profit:", total_profit)
print("Average Daily Profit:", round(total_profit / simulation_days, 2))

# Visualization
plt.plot(range(1, simulation_days + 1),
         daily_profit,
         marker='o')

plt.xlabel("Day")
plt.ylabel("Profit")
plt.title("Profit Analysis Simulation")
plt.grid(True)

plt.show()
```
