```python
# Question:
# Simulate the present and proposed bearing replacement policy
# for 10000 hours and choose the better policy.

import random

# Bearing life distribution
bearing_life = [1200, 1400, 1600, 1800, 2000]
bearing_prob = [0.10, 0.25, 0.30, 0.25, 0.10]

# Delay time distribution
delay_time = [2, 5, 8]
delay_prob = [0.3, 0.5, 0.2]

SIM_TIME = 10000  # hours


# ----------------------------
# Present Policy
# ----------------------------
def present_policy():

    time = 0
    total_cost = 0

    bearings = [
        random.choices(bearing_life, weights=bearing_prob)[0],
        random.choices(bearing_life, weights=bearing_prob)[0],
        random.choices(bearing_life, weights=bearing_prob)[0]
    ]

    while time < SIM_TIME:

        failure = min(bearings)
        time += failure

        if time >= SIM_TIME:
            break

        # reduce remaining life
        bearings = [x - failure for x in bearings]

        failed = bearings.index(0)

        delay = random.choices(delay_time,
                               weights=delay_prob)[0]

        downtime = delay + 15

        # costs
        total_cost += downtime * 5
        total_cost += (15 / 60) * 30
        total_cost += 30

        # replace failed bearing
        bearings[failed] = random.choices(
            bearing_life,
            weights=bearing_prob
        )[0]

    return total_cost


# ----------------------------
# Proposed Policy
# ----------------------------
def proposed_policy():

    time = 0
    total_cost = 0

    bearings = [
        random.choices(bearing_life, weights=bearing_prob)[0],
        random.choices(bearing_life, weights=bearing_prob)[0],
        random.choices(bearing_life, weights=bearing_prob)[0]
    ]

    while time < SIM_TIME:

        failure = min(bearings)
        time += failure

        if time >= SIM_TIME:
            break

        delay = random.choices(delay_time,
                               weights=delay_prob)[0]

        downtime = delay + 35

        # costs
        total_cost += downtime * 5
        total_cost += (35 / 60) * 30
        total_cost += 3 * 30

        # replace all bearings
        bearings = [
            random.choices(bearing_life,
                           weights=bearing_prob)[0]
            for _ in range(3)
        ]

    return total_cost


# Run simulation
present = present_policy()
proposed = proposed_policy()

print("Present Policy Cost : $", round(present, 2))
print("Proposed Policy Cost: $", round(proposed, 2))

if present < proposed:
    print("\nBetter Policy: Present Policy")
else:
    print("\nBetter Policy: Proposed Policy")
```
