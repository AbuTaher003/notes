```python
# Question:
# 9. Simulate a maintenance and replacement problem.

import random

# Machine life distribution
life = [1, 2, 3, 4, 5]
prob = [0.1, 0.2, 0.4, 0.2, 0.1]

simulation_time = 50

time = 0
replacement_count = 0
total_cost = 0

replacement_cost = 100
maintenance_cost = 20

while time < simulation_time:

    # Generate machine life
    machine_life = random.choices(life, weights=prob)[0]

    # Maintenance cost each cycle
    total_cost += maintenance_cost

    # If machine fails within simulation period
    if time + machine_life <= simulation_time:
        replacement_count += 1
        total_cost += replacement_cost

    time += machine_life

print("Simulation Time:", simulation_time)
print("Number of Replacements:", replacement_count)
print("Total Cost:", total_cost)
```

```python
import matplotlib.pyplot as plt

# Store replacement times
replacement_times = []

time = 0
replacement_count = 0
total_cost = 0

while time < simulation_time:

    machine_life = random.choices(life, weights=prob)[0]

    total_cost += maintenance_cost

    if time + machine_life <= simulation_time:
        replacement_count += 1
        replacement_times.append(time + machine_life)
        total_cost += replacement_cost

    time += machine_life

# Visualization
plt.figure(figsize=(8,4))
plt.bar(range(1, len(replacement_times)+1), replacement_times)

plt.xlabel("Replacement Number")
plt.ylabel("Time of Failure")
plt.title("Machine Replacement Timeline")
plt.grid(True)

plt.show()
```
