```python
# Question:
# Simulate an Inventory System.

import random
import matplotlib.pyplot as plt

# Parameters
inventory = 50          # Initial inventory
reorder_level = 20
reorder_quantity = 40

simulation_days = 30

inventory_history = []

# Simulation
for day in range(1, simulation_days + 1):

    # Random daily demand
    demand = random.randint(1, 10)

    inventory -= demand

    # Reorder if inventory reaches reorder level
    if inventory <= reorder_level:
        inventory += reorder_quantity
        print(f"Day {day}: Reordered {reorder_quantity} units")

    inventory_history.append(inventory)

print("\nFinal Inventory:", inventory)

# Visualization
plt.plot(range(1, simulation_days + 1),
         inventory_history,
         marker='o')

plt.xlabel("Day")
plt.ylabel("Inventory Level")
plt.title("Inventory System Simulation")
plt.grid(True)

plt.show()
```
