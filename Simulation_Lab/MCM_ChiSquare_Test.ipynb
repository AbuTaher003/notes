```python
# Question:
# You are tasked with evaluating the quality of a random number generator
# using the Multiplicative Congruential Method (MCM) and verifying its
# uniformity using the Chi-Square goodness-of-fit test.

import math

# Multiplicative Congruential Method parameters
seed = 7       # X0
a = 5          # multiplier
m = 16         # modulus
n = 20         # number of random numbers

random_numbers = []

# Generate random numbers
x = seed
for i in range(n):
    x = (a * x) % m
    random_numbers.append(x)

print("Generated Numbers:")
print(random_numbers)

# Convert to Uniform(0,1)
u = []
for x in random_numbers:
    u.append(x / m)

print("\nUniform Numbers:")
print(u)

# Chi-Square Test
k = 5                    # number of classes
expected = n / k         # expected frequency

freq = [0] * k

# Count frequencies
for x in u:
    index = int(x * k)
    if index == k:
        index = k - 1
    freq[index] += 1

print("\nClass\tObserved(O)\tExpected(E)\t(O-E)^2/E")

chi_square = 0

for i in range(k):
    value = ((freq[i] - expected) ** 2) / expected
    chi_square += value

    print(i+1, "\t",
          freq[i], "\t\t",
          expected, "\t\t",
          round(value, 4))

print("\nChi-Square Value =", round(chi_square, 4))

# Critical value for df=4 and alpha=0.05
critical = 9.488

if chi_square < critical:
    print("Accept H0 (Uniform Distribution)")
else:
    print("Reject H0 (Not Uniform Distribution)")
```
