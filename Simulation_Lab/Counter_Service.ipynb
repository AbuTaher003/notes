```python
# Question:
# Simulate a Counter Service.

import random
import matplotlib.pyplot as plt

# Parameters
simulation_time = 50

arrival_time = 0
customers_served = 0

waiting_times = []

# Simulation
while arrival_time < simulation_time:

    # Customer arrival interval
    inter_arrival = random.randint(1, 5)

    # Service time
    service_time = random.randint(2, 6)

    arrival_time += inter_arrival

    if arrival_time <= simulation_time:
        customers_served += 1
        waiting_times.append(service_time)

print("Total Customers Served:", customers_served)
print("Average Service Time:",
      round(sum(waiting_times)/len(waiting_times), 2))

# Visualization
plt.plot(
    range(1, customers_served + 1),
    waiting_times,
    marker='o'
)

plt.xlabel("Customer Number")
plt.ylabel("Service Time")
plt.title("Counter Service Simulation")
plt.grid(True)

plt.show()
```
